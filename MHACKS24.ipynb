```python
import zipfile
```

```python
import pandas as pd
```

```python
from chembl_webresource_client.new_client import new_client
```

```python
from io import StringIO
```

```python
# Target search for Ubiquitin receptors
target = new_client.target
target_query = target.search('Ubiquitin carboxyl-terminal hydrolase 1')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target)
df = pd.DataFrame.from_dict(res)
df
```

```python
csv_buffer = StringIO()
df.to_csv(csv_buffer, index=False)

with zipfile.ZipFile('data.zip', 'w', zipfile.ZIP_DEFLATED) as zipf:
    zipf.writestr('data.csv', csv_buffer.getvalue())
```

```python
df2 = df[df.standard_value.notna()]
df2 = df2[df.canonical_smiles.notna()]
df2
```

```python
len(df2.canonical_smiles.unique())
```

DROP DUPLICATE DATA

```python
df2_n1 = df2.drop_duplicates(['canonical_smiles'])
df2_n1
```

```python
bioactivity_class = []
for i in df2_n1.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df3 = df2_n1[selection]
df3
```

```python
data_tuples = list(zip(df3.molecule_chembl_id, df3.canonical_smiles, bioactivity_class, df3.standard_value))
df3_1 = pd.DataFrame(data_tuples, columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class','standard_value'])
df3_1
```

```python
df3_1.to_csv(csv_buffer, index=False)

with zipfile.ZipFile('ubiquitin_preprocessed.zip', 'w', zipfile.ZIP_DEFLATED) as zipf:
    zipf.writestr('ubiquitin_hydrolase_1_data.csv', csv_buffer.getvalue())
```
